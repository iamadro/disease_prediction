## User Input Prediction

```python
def Input_Prediction(given_symps):
    symps = list(x.columns)
    symps_dict = dict.fromkeys(symps, 0)
    # print("Before: ", symps_dict)

    # given_symps = ['skin_rash', 'nodal_skin_eruptions', 'dischromic_patches']
    for i in range (len(given_symps)):
        symps_dict[ given_symps[i]] =1

    lt = list (symps_dict.values()) 
    arr = np.array(lt)
    if( len(arr)>132 ):
        return -1
    arr= arr.reshape(1,-1)
    # print(arr.shape)
    # print(arr)
    return rfc.predict(arr)[0]
```

## Creating the coded_num : disease_name dict

```python
y=list(y)
u = []
for i in y :
    if i not in u:
        u.append(i)    

l = []
for i in y_:
    if i not in l:
        l.append(i)
        
dic = dict()

for i in range (len(u)):
    dic[ u[i] ] = l[i]
    
dic
```

### Prediction

```python
# print(Input_Prediction(['skin_rash', 'nodal_skin_eruptions', 'dischromic_patches']))
# given_symps = list (input("Enter the symptoms")

given_symps=['acidity', 'stomach_pain', 'ulcers_on_tongue']
predicted_coded_val = Input_Prediction(given_symps)
if (predicted_coded_val==-1):
    print("Please check the input symptoms")
else:
    print (dic[predicted_coded_val])
```
